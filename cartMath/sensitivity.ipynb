```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib
font = {'size'   : 20}
matplotlib.rc('font', **font)
```

```python
df = pd.read_csv("results/fig7_data.csv")
df.head(5)
```

```python
CM_CR = df[df["Response"]=="CR"]
CM_NR = df[df["Response"]=="NR"]
xvars = df.columns.to_list()[:-3]
CR55 = np.around(CM_CR.corr()["CM55"][:-2].values,decimals=3)
CR75 = np.around(CM_CR.corr()["CM75"][:-2].values,decimals=3)
NR55 = np.around(CM_NR.corr()["CM55"][:-2].values,decimals=3)
NR75 = np.around(CM_NR.corr()["CM75"][:-2].values,decimals=3)
```

```python
fig, ax = plt.subplots(2,2,figsize=(16,12))

bars = ax[0,0].barh(xvars, CR55,color="green")
ax[0,0].set_xlim([-0.5,0.5])
ax[0,0].grid("on")
for bars in ax[0,0].containers:
    ax[0,0].bar_label(bars)
ax[0,0].set_title("$C_M$ at t = 55 days (CR)")


bars = ax[0,1].barh(xvars, NR55,color="red")
ax[0,1].set_xlim([-0.5,0.5])
ax[0,1].grid("on")
for bars in ax[0,1].containers:
    ax[0,1].bar_label(bars)
ax[0,1].set_title("$C_M$ at t = 55 days (NR)")
    
bars = ax[1,0].barh(xvars, CR75,color="green")
ax[1,0].set_xlim([-0.5,0.5])
ax[1,0].grid("on")
for bars in ax[1,0].containers:
    ax[1,0].bar_label(bars)
ax[1,0].set_title("$C_M$ at t = 75 days (CR)")

    
bars = ax[1,1].barh(xvars, NR75,color="red")
ax[1,1].set_xlim([-0.5,0.5])
ax[1,1].grid("on")
for bars in ax[1,1].containers:
    ax[1,1].bar_label(bars)
ax[1,1].set_title("$C_M$ at t = 75 days (NR)")
    
plt.show()
```
